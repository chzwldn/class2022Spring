1. Autocorrelation

*   a calculated value representing **how similar a value within a time series is to a previous value**
*   correlation of a time series **with a lagged copy of itself**, in other words
*   x축은 lag의 개수, y축은 correlation으로 그려지며 **lag 0은 항상 1**일 수밖에 없다







2. Root Mean Square


*   RMS value of a signal is calculated as the **square root of average of squared value of the signal**

*   When two uncorrelated (or orthogonal ) signals are added together, such as noise from two independent sources, the RMS value of their sum is equal to the square-root of sum of the square of their individual RMS values.
